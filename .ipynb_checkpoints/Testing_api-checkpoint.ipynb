```python
#upload image
import PIL 
import numpy as np
import base64
import requests
import json
from PIL import Image
```

```python
#open image and convert to RGB
test_pic = Image.open('/Users/oscarrignell/Desktop/bermuda.png').convert('RGB') #pic in PIL
```

```python
#turn to array [0:255]
test_pic_array = np.array(test_pic)
```

```python
#switch to U-Int 8 (readable by bit64)
test_pic_array = test_pic_array.astype('uint8')
```

```python
#memorizing shape
height, width, channel = test_pic_array.shape
height, width, channel
```

```python
#reshape
test_pic_array = test_pic_array.reshape(height*width*channel)
test_pic_array
```

```python
# encoding to b64
b64bytes = base64.b64encode(test_pic_array)
```

```python
#decoding to utf8 and turning to  string
b64str = b64bytes.decode('utf8').replace("'", '"')
image_dict = {
    'image': b64str,
    'height': height,
    'width': width, 
    'channel': channel
}
```

```python
decoded = base64.b64decode(bytes(image_dict['image'], 'utf-8'))
decoded = np.frombuffer(decoded, dtype='uint8')
decoded = decoded.reshape(image_dict['height'], image_dict['width'], image_dict['channel'])
print(decoded.shape)
```

```python
# image dict contains info on image, height, width and channels
```

```python
headers= {
    'Content_Type': 'application/json'
}
requests.post('http://127.0.0.1:8000/predict/', json.dumps(image_dict), headers=headers)
#fastAPI.py POST code to add on VS Code
```
